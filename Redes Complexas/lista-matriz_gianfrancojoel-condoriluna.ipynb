```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Função para converter lista de adjacência em matriz de adjacência
def lista_para_matriz(lista_adj, n):
    matriz = np.zeros((n, n), dtype=int)
    for i, vizinhos in lista_adj.items():
        for vizinho in vizinhos:
            matriz[i][vizinho] = 1
            matriz[vizinho][i] = 1  # Porque é um grafo não direcionado
    return matriz

# Função para converter matriz de adjacência em lista de adjacência
def matriz_para_lista(matriz):
    lista_adj = {i: [] for i in range(len(matriz))}
    for i in range(len(matriz)):
        for j in range(len(matriz)):
            if matriz[i][j] == 1:
                lista_adj[i].append(j)
    return lista_adj

# Exemplo usando o grafo do clube de karatê de Zachary
# Carregar o grafo do clube de karatê de Zachary
G = nx.karate_club_graph()

# Obter a lista de adjacência do grafo
lista_adj = nx.to_dict_of_lists(G)

# Converter a lista de adjacência em matriz de adjacência
matriz_adj = lista_para_matriz(lista_adj, len(G.nodes))

# Imprimir a matriz de adjacência
print("Matriz de Adjacência:")
print(matriz_adj)

# Converter a matriz de adjacência de volta para lista de adjacência
nova_lista_adj = matriz_para_lista(matriz_adj)

# Imprimir a lista de adjacência
print("\nLista de Adjacência:")
for k, v in nova_lista_adj.items():
    print(f"{k}: {v}")

# ------------------------------------------------------------
# Agora, visualizando o grafo do clube de karatê de Zachary
# ------------------------------------------------------------

# Posição dos nós para uma visualização mais clara
pos = nx.spring_layout(G)

# Obter os grupos a partir do atributo 'club' (Mr. Hi's club e o clube do Administrador)
grupo_1 = [n for n in G.nodes if G.nodes[n]['club'] == 'Mr. Hi']
grupo_2 = [n for n in G.nodes if G.nodes[n]['club'] == 'Officer']

# Desenhar os nós do grupo 1 (Mr. Hi) em azul claro
nx.draw_networkx_nodes(G, pos, nodelist=grupo_1, node_color='lightblue', label='Mr. Hi')

# Desenhar os nós do grupo 2 (Administrador) em verde claro
nx.draw_networkx_nodes(G, pos, nodelist=grupo_2, node_color='lightgreen', label='Officer')

# Desenhar as arestas (conexões entre membros)
nx.draw_networkx_edges(G, pos)

# Desenhar os rótulos dos nós
nx.draw_networkx_labels(G, pos)

# Mostrar a legenda e o gráfico
plt.legend(['Mr. Hi', 'Officer'])
plt.title("Grafo do Clube de Karatê de Zachary")
plt.show()

```
